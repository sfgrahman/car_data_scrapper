```python
import re


def find_brands_in_sentence(sentence, brand_name):
    matches = []
    for brand in brand_name:
        # Find all occurrences of the brand in the sentence
        match = re.findall(rf"\b{re.escape(brand)}\b", sentence)
        if match:
            matches.extend(match)  # Add the matched brand to the list
    return matches[0] if matches else ""
```

```python
total = []
for sentence in sentences:
    if isinstance(sentence, str):
        # Call the function and print the result
        result = find_brands_in_sentence(sentence, brand_name)
        total.append(result)
    else:
        print(sentence)
        total.append("")
```

```python
len(total)
```

```python
data["Make"] = total
```

```python
pattern = r"\b\d{4}\b"
final_year = []
for sentence in sentences:
    match = re.search(pattern, sentence)

    # Extracted year
    if match:
        #print(match.group(0))
        year = match.group(0)
    else:
        year = ""
    final_year.append(year)
```

```python
len(final_year)
```

```python
data["Year"] = final_year
```

```python
test_d = data[["Title_sep", "Make"]]
```

```python
model_final = []
for idx, (Title, Make) in enumerate(zip(test_d["Title_sep"], test_d["Make"])):
    if Make:
        pattern = rf"\b{re.escape(Make)}\b"
        parts = re.split(pattern, Title)
        model_final.append(parts[1])
    else:
        model_final.append("")
```
